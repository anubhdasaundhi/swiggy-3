```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
df_top_10['city_list'] = df['city'].apply(
    lambda x: [i.strip() for i in x.split(',')]
)
```

```python
df_top_10['cuisine_list'] = df_top_10['cuisine'].apply(
    lambda x: [i.strip() for i in x.split(',')]
)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb_city = MultiLabelBinarizer()
city_encoded = mlb_city.fit_transform(df_top_10['city_list'])

city_df = pd.DataFrame(
    city_encoded,
    columns=['city_' + c for c in mlb_city.classes_],
    index=df_top_10.index
)
```

```python
mlb_cuisine = MultiLabelBinarizer()
cuisine_encoded = mlb_cuisine.fit_transform(df_top_10['cuisine_list'])

cuisine_df = pd.DataFrame(
    cuisine_encoded,
    columns=['cuisine_' + c for c in mlb_cuisine.classes_],
    index=df_top_10.index
)
```

```python







df_final = pd.concat([df_top_10, city_df, cuisine_df], axis=1)

df_final.drop(
    columns=['city', 'cuisine', 'city_list', 'cuisine_list'],
    inplace=True
)
```

```python
import matplotlib.pylab as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection  import train_test_split
from sklearn.cluster import KMeans
```

```python
df2=df_final.iloc[:,1:]
```

```python
from scipy.spatial.distance import cdist
clusters=range(1,20)
meanDistortions=[]

for k in clusters:
    model=KMeans(n_clusters=k)
    model.fit(df2)
    prediction=model.predict(df2)
    meanDistortions.append(sum(np.min(cdist(df2, model.cluster_centers_, 'euclidean'), axis=1)) / df2.shape[0])


plt.plot(clusters, meanDistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method')
```

```python
df2.head()
```

```python
final_model=KMeans(20)
final_model.fit(df2)
prediction=final_model.predict(df2)

#Append the prediction 
df_final["GROUP"] = prediction
df2["GROUP"] = prediction
print("Groups Assigned : \n")
df_final.head()
```

```python
import pickle  # Add at top if not already imported

# Save the trained KMeans model
with open('kmeans_swiggy_model.pkl', 'wb') as f:
    pickle.dump(final_model, f)
```
